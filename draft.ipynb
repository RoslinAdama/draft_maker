### Création au hasard de compos

```python
from random import choice
from tqdm import tqdm

if "Unnamed: 0" in joueurs.keys():
    del joueurs["Unnamed: 0"]

compos_random = []

for i in tqdm(range(50000)):
    compo = []
    for j_potentiels in joueurs.values():
        joueur_choisi = choice(j_potentiels)
        compo.append(joueur_choisi)

    if compo not in compos_random:
        compos_random.append(compo)

compos_random
```

### Filtrage par équipes

```python
compos_valides = []

def find_equipe(joueur):
    index_equipe = df_noms[df_noms.values == joueur].index[0]
    return df_noms.at[index_equipe, "Unnamed: 0"]

for compo in tqdm(compos_random):
    equipes = [find_equipe(joueur) for joueur in compo]
    # print(equipes)
    if not any([equipes.count(value)>1 for value in equipes]):
        compos_valides.append(compo)

compos_valides
```

### Lien entre df_noms et df_notes

```python
import numpy as np

def notes_from_nom(nom_joueur : str) -> float:

    result_search = np.where(df_noms == nom_joueur)
    num_ligne = result_search[0][0]
    num_col = result_search[1][0]

    return df_notes.iloc[num_ligne, num_col]

notes_from_nom("L Martinez")
```

```python
from typing import List

def valeur_equipe(equipe : List[str]) -> float:

    valeurs = [notes_from_nom(joueur) for joueur in equipe]

    return sum(valeurs)

valeurs = [ valeur_equipe(compo) for compo in compos_valides ]

meilleure_compo = compos_valides[valeurs.index(max(valeurs))]

meilleure_compo
```

```python
def get_equipe_from_joueur(nom_joueur : str):

    result_search = np.where(df_noms == nom_joueur)

    num_ligne = result_search[0][0]

    return df_noms["Unnamed: 0"][num_ligne]

get_equipe_from_joueur("A Laurienté")

    
```
